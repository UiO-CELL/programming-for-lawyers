```python
# Task 2.1 part 1
# Takes a filename as a parameter, opens the file using pandas read_csv(), returns the resulting DataFrame
from pandas import pandas

def open_file(filename):
    try:
        return pandas.read_csv(filename, index_col='docname')
    except:
        print("Something went wrong")
```

```python
# Task 2.1 part 2
# Assigns the file 'cases-10.csv' to a variable 'cases' and displays the head (first 5 rows) of the table. (I imagine you by "object", mean variable, even though a variable isn't an object itself.)
cases = open_file('/src/data/JUS5080/cases-10.csv')
display(cases.head())
```

```python
# Task 2.2
# Selects cases (rows) where article 8 is relevant (column "article=8" contains 1)
article_8_cases = cases.loc[cases['article=8'] == 1]
display(article_8_cases)
```

```python
# Task 2.3
# Selects only the columns the have names starting with 'article'
selection = cases.filter(regex='^article=p')
selection.sum(axis=0).plot(kind='bar')
```

```python
# Task 2.4
# Makes a plot of the number of non-violations for each article

#Retrieves cases
cases_2000 = open_file('/src/data/JUS5080/cases-2000.csv')

# Selects columns starting with 'ccl_article'
selection = cases_2000.filter(regex='^ccl_article')

# Removes entries signifying violations (by turning 1s to 0s)
data = selection.replace(1, 0)

# Makes the bars go upwards by replacing -1s with 1s
data = data.replace(-1, 1)

# Sums the columns
data = data.sum(axis=0)

# Plots the results as a bar chart
data.plot(kind='bar', figsize=(12,5))
```

```python
# Task 2.5
# Makes a bar chart for the number of decisions each year

# Parses the 'decisiondate' column in cases_2000 into machine readable format (Strings are already machine readable by the way)
cases_2000['decisiondate'] = pandas.to_datetime(cases_2000['decisiondate'], dayfirst=True, format="%d/%m/%Y")

# Gets the year part of the dates 
data = cases_2000['decisiondate'].dt.year

# Finds the number of decisions each year using the value_counts function
data = data.value_counts()

# Sorts the results by year
data = data.sort_index()

# Plots the results in a bar chart
data.plot(kind='bar')
```

```python
# Task 2.6
# Takes two arguments: a pandas DataFrame of cases, and an article number (int)
# Prints the article number and the maximum number of violations for the article (I assume you mean the maximum in a case, and not the total accross cases)
# Returns a pandas DataFrame with the case(s) having the highest number of violations (I assume having the exact maximum is what is meant)
def get_cases_with_most_article_violations(cases, article_number):
    # Prints the article number
    print("Article number: " + str(article_number))
    # Makes a selection of the columns for the article
    selection = cases.filter(regex='^ccl_article=' + str(article_number))
    # Removes entries signifying non-violations (by turning -1s to 0s)
    data = selection.replace(-1, 0)
    # Sums violations on the column axis
    data = data.sum(axis=1)
    # Prints maximum amount of violations
    maximum = data.max()
    print("Maximum number of violations: " + str(maximum))
    # Returns the case(s) having the (exact) highest number of violations
    data = data.to_frame(name='violations')
    data = data.loc[data['violations'] == maximum]
    return data

display(get_cases_with_most_article_violations(cases_2000, 5))
```
